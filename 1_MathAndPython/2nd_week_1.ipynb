```python
import numpy as np
import pandas as pd
import re
```

```python
f = open('sentences.txt')
```

```python
lines = []
for line in f:
    lines.append(line.lower())
```

```python
tokens = {}
counter = 0
locals = []

for line in lines:
    local_tokens = re.split('[^a-z]', line)
    local = {}
    for token in local_tokens:
        
        if token not in tokens:
            tokens[token] = counter
            counter = counter+1
            
        if token in local:
            local[token] = local[token]+1
        elif token!='':
            local[token]=1
            
    locals.append(local)
```

```python
len(tokens) #d = 254
```

```python
for sentence in locals:
    resultString = []
    
    for word in sentence:
        resultString.append(tokens[word])
        
    print(resultString)
```
